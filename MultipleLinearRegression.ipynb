```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independant=dataset[['R&D Spend']]
dependant=dataset[['Profit']]
```

```python
plt.scatter(independant,dependant,color='grey')
plt.xlabel('Administration', fontsize=15)
plt.ylabel('Profit', fontsize=15)
plt.show()
```
